## 1. Load the data and split the data:

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
# To-Do:  load the data 
df = pd.read_csv("clean_HepatitisC.csv")
df.head()
```

```python
# To-Do : keep only the columns to be used : all features except ID, cluster 
# The target here is the Category column 
# Do not forget to split your data (this is a classification task)
# test set size should be 20% of the data 

df1 = df.drop(columns=["ID","cluster"])
y = df1["Category"]
x = df1.drop(columns="Category")

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=50)
```

## 2. Building the model and training and evaluate the performance: 

```python
# To-do build the model and train it 
# note that you will be providing explanation about the hyper parameter tuning 
# So you will be iterating a number of times before getting the desired performance 

model = RandomForestClassifier(criterion = "entropy" ,n_estimators = 50 , max_features = 12 , max_leaf_nodes = 5 , 
                               bootstrap = True , min_samples_split = 4 ,class_weight = "balanced",
                               min_samples_leaf = 2 , n_jobs = -1 , max_samples = 124 , random_state = 16180)

model.fit(x, y) 

# gini gives poor results when it comes to Hepatitis and Fibrosis class due to the small size of elements in the its training set 
# max_features was set to the number of features we have 
# max_leaf_nodes was set to the number of classes we have since the max number of samples in each tree was set to 124 > max size of any class in the training set
# class_weight = balanced because 86% of the whole dataset is of class 0 
```

```python
y_hat = model.predict(x_test)
```

```python
# To-do : evaluate the model in terms of accuracy and precision 
# Provide evidence that your model is not overfitting 
from sklearn.metrics import precision_score, accuracy_score

#classification_report
print("Precision score :%.4f\n"%precision_score(y_test,y_hat,average="weighted"))
print("the accuracy : %.4f "%accuracy_score(y_test,y_hat))
```

In this part you need to add the predicted class as a column to your test dataframe and save this one
